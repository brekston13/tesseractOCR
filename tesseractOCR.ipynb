```python
# Install Tesseract OCR
!apt-get install -y tesseract-ocr
# Install the necessary Python libraries
!pip install pytesseract opencv-python

import cv2
import pytesseract
from google.colab import files
import matplotlib.pyplot as plt

# Step 1: Upload an image file
uploaded = files.upload()
image_path = list(uploaded.keys())[0]

# Step 2: Read the image using OpenCV
image = cv2.imread(image_path)

# Step 3: Convert to grayscale
gray = cv2.cvtColor(image, cv2.COLOR_BGR2GRAY)

# Step 4: Apply preprocessing (optional)
gray = cv2.medianBlur(gray, 3)  # Optional: You can experiment with other preprocessing techniques

# Show the preprocessed image
plt.imshow(gray, cmap='gray')
plt.axis('off')
plt.show()

# Step 5: Perform OCR on the preprocessed image
custom_config = r'--oem 3 --psm 6'  # You can adjust these parameters as needed
text = pytesseract.image_to_string(gray, config=custom_config)

# Step 6: Print the extracted text
print("Extracted Text:")
print(text)

# Purpose of the Script:
# This script demonstrates a simple implementation of Optical Character Recognition (OCR) using Tesseract in Python.
# It allows users to upload an image, preprocess it, and extract text content from it.
# OCR has a wide range of applications, including digitizing printed documents, automating data entry, and enhancing accessibility for visually impaired individuals.
# By using this code, developers can integrate OCR capabilities into their applications, improving user interaction and data processing efficiency.
```
